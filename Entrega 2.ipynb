```python
##Ahora voy a realizar un gráfico de líneas utilizando dos columnas.


#Preparo los datos para el gráfico
x = df["empresa"].head(20)
y = df["ï»¿indice_tiempo"].head(20)

# Creo la figura y los ejes
fig, ax = plt.subplots()

# Creo el gráfico de líneas
ax.plot(x, y)

# Configuro los ejes y el título del gráfico
ax.set_xlabel("Empresa")
ax.set_ylabel("Fecha")
ax.set_title("Gráfico de Datos")

# Muestro el gráfico
plt.show()
```

```python
##Perfecto. Ahora algo más complejo: Un gráfico de barras. Sería interesante ver la cantidad de veces que aparece una empresa y hace una compra. Para éso, primero debería filtrar cuantas empresas distintas hay y luego ver la frecuencia en la que c/u realizó una operación.

# Filtro valores únicos y cuento frecuencias
valores_unicos = df['empresa'].unique()
frecuencias = df['empresa'].value_counts()

# Tomo solo los primeros 7 valores para el gráfico
primeros_valores = valores_unicos[:7]
primeras_frecuencias = frecuencias[:7]

# Creo una figura y un eje
fig, ax = plt.subplots()

# Creo el gráfico de barras utilizando el eje
ax.bar(primeros_valores, primeras_frecuencias, width=0.55)

# Establezco etiquetas y título
ax.set_xlabel('Empresa')
ax.set_ylabel('Cantidad de compras')
ax.set_title('Gráfico de barras de cantidad de compras por empresa')

# Roto las etiquetas del eje x para que no se encimen.
ax.set_xticklabels(primeros_valores, rotation=45)

# Muestro el gráfico
plt.show()
```

```python
##Maravilloso! Histograma hecho! Ahora... hagamos un scatterplot. Supongamos que necesitamos saber qué empresas son las que menos compras realizaron. Pues bien, en lugar de seleccionar los primeros 7 valores, que son los que tienen más compras, ya que así están ordenados. Seleccionaríamos los últimos 7. Y como sabemos que los valore serán pequeños, es más estético utilizar un scatterplot. Veamos...

# Filtro valores únicos y cuento frecuencias
valores_unicos = df['empresa'].unique()
frecuencias = df['empresa'].value_counts()

# Tomo solo los primeros 7 valores para el gráfico
ultimos_valores = valores_unicos[-7:]
ultimas_frecuencias = frecuencias[-7:]

# Creo una figura y un eje
fig, ax = plt.subplots()

# Creao el gráfico de barras utilizando el eje
ax.scatter(ultimos_valores, ultimas_frecuencias)

# Establezco etiquetas y título
ax.set_xlabel('Empresa')
ax.set_ylabel('Cantidad de compras')
ax.set_title('Gráfico de barras cantidad de compras por empresa')

# Roto las etiquetas del eje x para que no se encimen.
ax.set_xticklabels(ultimos_valores, rotation=45)

# Muestro el gráfico
plt.show()
```

```python
#Qué interesante! Realizando éste ejercicio nos dimos cuenta de que no sólo existen empresas en el datafrane, sino también compras particulares. Casi planeado! Teniendo ésta información, ahora sabemos que se debería cambiar el nombre del eje X al de "Empresa/Particular". 
```

```python
#Pero ésta no es la principal función del gráfico de puntos, o scatterplot. Sino mostrar la relación entre dos o más variables, lo cual no fué propiamente logrado en el ejemplo anterior. Así que se me ocurre una mejor forma de utilizarlo: Vamos a filtrar otra variable del dataframe, la variable precio.

precios_unicos = df['precio'].unique() #Primero, vamos a separar las compras en los precios que más aparecen para clasificarlos de alguna manera.
frecuencia_de_precio = df['precio'].value_counts() #Ahora, vamos a ver con qué frecuencia se realizan compras en cada uno de ellos. De ésta forma, determinaremos en qué rango de precios hay más compras dentro de las 390mil.

# Creo una figura y un eje
fig, ax = plt.subplots()

# Creo el gráfico de barras utilizando el eje
ax.scatter(precios_unicos, frecuencia_de_precio)

# Establezco etiquetas y título
ax.set_xlabel('Cantidad de compras')
ax.set_ylabel('Precio')
ax.set_title('Scatterplot de frecuencia de precios')

# Muestro el gráfico
plt.show()
```

```python
#Una ventaja de utilizar scatterplot en éste caso fue que me permitió analizar con mayor claridad grandes volúmenes de información sin perder visualización. Como se puede observar, la mayor cantidad de compras se realizaron entre el rango de precios de 10 a 50, aproximadamente. De esta forma se realizó un correcto uso del scatterplot y pudimos analizar la relación (muy fuerte en este caso) entre el precio y la frecuencia de compra.
```

```python
#Sigamos con Seaborn!
import seaborn as sns
```

```python
#Seaborn facilita bastante la visualización. 

df_reducido = df.apply(lambda x: x.head(20)) #Achico el dataframe a 20 valores para realizar un muestreo claro. A continuación, realizaré un histograma con Seabron, se verá la mínima cantidad de código requerida para mostrar información valiosa.

sns.histplot(data= df_reducido, x= df_reducido['precio'], hue=df_reducido['empresa'], multiple="stack")
 
```

```python
#Como se visualiza, de esta manera somos capaces de obtener de manera sencilla y gráfica, la cantidad de compras a un determinado precio que las distintas empresas fueron realizando. Incluso podríamos especificar por empresa utilizando múltiples figuras de la siguiente manera:

sns.displot(data=df_reducido, x= df_reducido['precio'],  col="empresa")
```

```python
##Perfecto. Ahora veamos como sería un scatterplot de matplotlib hecho en seaborn. Para ésto, tomaré una variable adiciónal, la localidad. Esta vez visualizaremos con diferentes colores, las empresas que realizaron una operación, de dónde proceden y a qué precio fueron realizadas.
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
sns.scatterplot(data=df_reducido, x=df_reducido['precio'], y="empresa", hue="localidad", ax=axs[0])

sns.histplot(data=df_reducido, x="precio", hue="empresa", legend=False, ax=axs[1], multiple="stack")
fig.tight_layout()
```

```python
#Podemos sacar múltiple información. Por ejmplo, del gráfico se se concluye que la empresa que realiza las compras más variadas es "10 de setiembre s.a", mientras que las demas, compran por encima de los 180 de precio, aproximadamente.

#Hagamos un gráfico de barras nuevo que muestre los índices de tiempo en los que las distintas empresas compraron y el precio al que lo hicieron.
df_reducido2 = df.apply(lambda x: x.head(30))
ax = sns.barplot(x="empresa", y="precio", hue="ï»¿indice_tiempo", data= df_reducido2)
```
